```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.figure_factory as FF

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ExcelFormattedGISTEMPData2CSV.csv')
```

```python
trace1 = go.Scatter(x=df['Year'], y=df['Glob'], line=dict(color=('rgb(255,0,0)'), width=2), name='Global')
```

```python
trace2 = go.Scatter(x=df['Year'], y=df['NHem'], line=dict(color=('rgb(0,0,255)'), width=2), name='North Hemisphere')
```

```python
trace3 = go.Scatter(x=df['Year'], y=df['SHem'], line=dict(color=('rgb(0,255,0)'), width=2), name='South Hemisphere')
```

```python
layout = go.Layout(title='GISS Surface Temperature from 1880 to 2014', 
                   xaxis=dict(title='Year'),
                   yaxis=dict(title='Deviation'),
                   plot_bgcolor='rgb(255,255,255)')
```

```python
fig = go.Figure(data=[trace1, trace2, trace3], layout=layout)
```

```python
iplot(fig)
```
